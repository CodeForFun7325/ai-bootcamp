```python
titles = pd.DataFrame(bbc_data['title'])
```

# Process Data

```python
# lowercase titles
titles['title_lower'] = titles['title'].apply(lambda x: x.lower())
```

```python
# stop word removal
eng_stopwords = stopwords.words("english")
titles["title_nostop"] = titles["title_lower"].apply(lambda x: " ".join([word for word in x.split() if word not in (eng_stopwords)]))
titles.head()
```

```python
# punctuation removal
titles["title_nopunct_nostop"] = titles["title_nostop"].apply(lambda x: re.sub(r"[^\w\s]", "", x))
titles["title_nopunct_nostop"].head(10)
```

```python
# tokenize raw data and clean data
titles["token_raw"] = titles["title"].apply(lambda x: word_tokenize(x))
titles["token_clean"] = titles["title_nopunct_nostop"].apply(lambda x: word_tokenize(x))

titles.head()
```

```python
#lemmatizing
lemmatizer = WordNetLemmatizer()

titles["tokens_clean_lemmatized"] = titles["token_clean"].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
titles.head()
```

```python
# Create lists for just our tokens
tokens_raw_list = sum(titles["token_raw"], []) # unpack our lists of tokens into a single list
tokens_clean_list = sum(titles["tokens_clean_lemmatized"], [])
```

# POS Tagging

```python
nlp = spacy.load("en_core_web_sm")
```

```python
spacy_doc = nlp(" ".join(tokens_raw_list))
spacy_doc_clean = nlp(" ".join(tokens_clean_list))
```

```python
pos_df = pd.DataFrame(columns = ["token", "pos_tag"])
pos_df_clean = pd.DataFrame(columns = ["token", "pos_tag"])
```

```python
for token in spacy_doc: 
    pos_df = pd.concat(
        [
            pos_df, 
            pd.DataFrame.from_records(
                [
                    {
                        "token": token.text, 
                        "pos_tag": token.pos_
                    }
                ]
            )
        ], 
        ignore_index = True
    )

for token in spacy_doc_clean: 
    pos_df_clean = pd.concat(
        [
            pos_df_clean, 
            pd.DataFrame.from_records(
                [ 
                    {
                        "token": token.text, 
                        "pos_tag": token.pos_
                    }
                ]
            )
        ]
    )
```

```python
pos_df_counts = pos_df.groupby(["token", "pos_tag"]).size().reset_index(name="counts").sort_values(by="counts", ascending=False)
pos_df_counts_clean = pos_df_clean.groupby(["token", "pos_tag"]).size().reset_index(name="counts").sort_values(by="counts", ascending=False)
```

# NER

```python
ner_df = pd.DataFrame(columns=["token", "ner_tag"])

for token in spacy_doc.ents: 
    if pd.isna(token.label_) is False:
        ner_df = pd.concat([
            ner_df, pd.DataFrame.from_records(
                [{"token": token.text, "ner_tag": token.label_}]),
        ], ignore_index = True)
```
